## Step 1: Imports

```python
import pandas as pd
import datetime
import os
```

## Step 2: Set Variables

```python
cwd = os.getcwd()
print(f'File Path: {cwd}')

folder = '/folder' # Your folder name here

data_path = cwd + folder
print(f'Data Location: {data_path}')
```

## Step 3: Get all folders and files

```python
folder_files = os.listdir(data_path)
for f in folder_files:
    print(f)
```

## Step 4: Take a peek of data

```python
df_peek = pd.read_csv(data_path + '/file_name.csv') # read csv file

df_peek.head()
```

```python
df_peek2 = pd.read_json(data_path + '/file_name.jsonl') # read jsonl file

df_peek2.head()
```

## Step 5: Combine all files (csvs + jsonls)

```python
df_list = []
for file in folder_files:
    print(f'Reading {file}')
    file_path = data_path + '/' + file

    if file.endswith('.csv'):
        temp_df = pd.read_csv(file_path)
    elif file.endswith('.jsonl'):
        temp_df = pd.read_json(file_path)

    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)
print('Completed Dataframe Concatenation')
```

```python
df.info()
```

## Step 6: Save Bronze Layer

```python
df.to_parquet('/bronze.parquet', index = False) # save the parquet to the target folder
print("âœ… Cleaned data saved to Bronze layer")
```
